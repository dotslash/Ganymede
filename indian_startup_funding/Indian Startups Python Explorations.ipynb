## Misc code and playing around

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
```

```python
# Misc. utils.

import math
def str_to_num(usd_value):
    float_val = float(str(usd_value).replace(",",""))
    return float_val

# display floats so that humans can easily read.
def display_float(number):
    if (number > 1000):
        number = int(number)
    return "{:,}".format(number)
pd.set_option('display.float_format', lambda x: display_float(x))


def label_bar_plot(bar):
    # https://web.archive.org/web/20170228011551/http://matplotlib.org/examples/api/barchart_demo.html
    for _,rect in enumerate(bar.patches):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height,
                '%d' % int(height),
                ha='center', va='bottom')
        
# Splits labels to multiple into lines of width <= wrap_size.
def wrap_labels(ax, wrap_size):
    from textwrap import wrap
    labels = ax.get_xticklabels()
    ax.set_xticklabels(['\n'.join(wrap(x.get_text(), wrap_size)) for x in labels])
```

# Clean up data

## AmountInUSD  should be a number

```python
# Add AmountInUSD-proc which represents the numerical value of AmountInUSD.
df_startups["AmountInUSD-proc"] = df_startups["AmountInUSD"].apply(str_to_num)
df_startups["AmountInUSD-proc"].describe()
```

## Cleanup invstment types
We notice that some of the names are non standardized. The following pairs should be the same.
- Seed Funding = SeedFunding 
- PrivateEquity = Private Equity
- Crowd funding = Crowd Funding

```python
InvestmentTypeVC = df_startups.InvestmentType.value_counts()
print("Description of Investiment Types: ")
print(InvestmentTypeVC)
```

```python
def clean_investment_types(investment_type):
    names_to_standardize = {'SeedFunding': 'Seed Funding',
                            'PrivateEquity':'Private Equity',
                            'Crowd funding': 'Crowd Funding'}
    return names_to_standardize.get(investment_type, investment_type)
    
df_startups['InvestmentType'] = df_startups['InvestmentType'].apply(clean_investment_types) 
print(df_startups.InvestmentType.value_counts())
```

## Clean up investors data

```python
print("Investors:")
print(df_startups.InvestorsName.value_counts()[:15])
```

```python
import stringcase
# "Undisclosed investors", "UndisclosedInvestors", "undisclosed Investors" -> "Undisclosed Investors"
def standardize_investors(investor):
    if type(investor) == float and math.isnan(investor):
        return investor
    standardized = stringcase.titlecase(investor.title().replace(' ', ''))
    if "Undisclosed" in standardized:
        return "Undisclosed Investors"
    return standardized

df_startups["InvestorsName-proc"] = df_startups["InvestorsName"].apply(standardize_investors)
print(df_startups["InvestorsName-proc"].value_counts()[:15])
```

# Plots
## Plot about investment amount

```python

fig = plt.figure(figsize=(8,5))

# Plot the log values, but the x axis will display the actual value.
g = sns.distplot(np.log(df_startups["AmountInUSD-proc"] + 1).dropna())
g.set_xlabel('Investment amounts in USD', fontsize=12)
g.set_title("Hist of investment amounts in USD", fontsize=16)

# canvas should be drawn before getting xtick labels.
fig.canvas.draw()

# a => human_readable(e^a)
def log_to_normal(text):
    if text == '' :
        return ''
    log_val = float(text)
    ret_val = int(math.exp(log_val))
    # round off to nearest 1000.
    return display_float(1000*(int(ret_val/1000)))
    

new_labels = [log_to_normal(x.get_text()) for x in g.get_xticklabels()]
g.set_xticklabels(new_labels, rotation=90)

plt.show()
```

```python
plt.figure(figsize = (15,15))
plt.subplot(221)

it_graph = sns.countplot(x="InvestmentType", data=df_startups)
it_graph.set_xticklabels(it_graph.get_xticklabels(),rotation=45)
it_graph.set_title("Investiment Type count", fontsize=15)
it_graph.set_xlabel("investment Types", fontsize=12)
it_graph.set_ylabel("Count", fontsize=12)


plt.subplot(222)
plt.yscale('log')
amount_graph = sns.boxplot(x="InvestmentType", y="AmountInUSD-proc", 
            data=df_startups[df_startups['AmountInUSD-proc'] < 100000000])
amount_graph.set_xticklabels(amount_graph.get_xticklabels(), rotation=45)
amount_graph.set_title("Investment Types less than 1 bi", fontsize=15)
amount_graph.set_xlabel("Investment Types", fontsize=12)
amount_graph.set_ylabel("Amount(USD)", fontsize=10)
import matplotlib.ticker as mtick
amount_graph.yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))


plt.show()
```

# Investors by vertical

```python
def DrawWordCloud(series, title):
    plt.figure(figsize=(15, 15))

    stopwords = set(STOPWORDS)
    import wordcloud
    wordcloud = wordcloud.WordCloud(
        background_color='black',
        stopwords=stopwords,
        max_words=150,
        max_font_size=40,
        random_state=42
    ).generate(str(series))

    print(wordcloud)
    fig = plt.figure(1)
    plt.imshow(wordcloud)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
consumer_investors = df_startups[df_startups['IndustryVertical'] == 'Consumer Internet']['InvestorsName']
tech_investors = df_startups[df_startups['IndustryVertical'] == 'Technology']['InvestorsName']
ecommerce_investors = df_startups[df_startups['IndustryVertical'] == 'Ecommerce']['InvestorsName']


DrawWordCloud(consumer_investors, 'Consumer Internet Investors')
DrawWordCloud(tech_investors, 'Tech Investors')
DrawWordCloud(ecommerce_investors, 'Ecommerce Investors')
```

# Investments over time

```python
df_startups.Date.replace((['12/05.2015', '13/04.2015','15/01.2015','22/01//2015']), \
                         ('12/05/2015','13/04/2015','15/01/2015','22/01/2015'), inplace=True)
```

```python
df_startups['Date'] = pd.to_datetime(df_startups['Date'])

df_startups['Date_month_year'] = df_startups['Date'].dt.to_period("M")
df_startups['Date_year'] = df_startups['Date'].dt.to_period("A")
```

```python
plt.figure(figsize=(14,6))

month_year_invesments = df_startups['Date_month_year'].sort_values()
sns.countplot(month_year_invesments)
plt.xticks(rotation=90)
plt.xlabel('Dates', fontsize=12)
plt.ylabel('Date Counting', fontsize=12)
plt.title("Count frequency Investiments Date ", fontsize=16)

plt.show()
```
